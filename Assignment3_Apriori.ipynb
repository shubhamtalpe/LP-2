```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset = pd.read_csv("store_data.csv", header=None)
```

```python
dataset.head()
```

```python
transactions = []
for i in range(dataset.shape[0]):
    transac = []
    for j in range(dataset.shape[1]):
        if(dataset.values[i, j] != dataset.values[i, j]):
            break
        else:
            transac.append(dataset.values[i, j])
    transactions.append(transac)
```

```python
transac_encoder = TransactionEncoder()
te_ary = transac_encoder.fit(transactions).transform(transactions, sparse=True)
df = pd.DataFrame.sparse.from_spmatrix(te_ary, columns=transac_encoder.columns_)
df
```

```python
df = df.sample(n=3000)
df
```

```python
frequent_items = apriori(df, min_support = 0.01, use_colnames=True)
```

```python
frequent_items
```

```python
frequent_items = apriori(df, min_support = 0.01, use_colnames=True)
frequent_items['length'] = frequent_items['itemsets'].apply(lambda x: len(x))
frequent_items
```

```python
frequent_items[frequent_items['itemsets'] == {'milk'}]
```

```python
rules = association_rules(frequent_items, metric="lift", min_threshold = 1)
print(rules)
```
